```python
def get_bounding_box(geojson: dict, outcode: str):

    newcoordinates = pd.Series()

    for i, item in enumerate(geojson):
            ditem = json.loads(item)
            print("{} : {}, {}".format(i, ditem['properties']['name'], ditem['properties']['styleUrl']))
            print("{}".format(len(ditem['geometry']['coordinates'][0])))
            newcoordinates = newcoordinates.append(pd.Series(ditem['geometry']['coordinates'][0]))

    newcoordinates.reset_index(drop=True, inplace=True)
    uniquecoordinates = newcoordinates.loc[newcoordinates.astype(str).drop_duplicates(keep = False).index]
    
    return(uniquecoordinates, str)
```

```python
pdTFL_tubes.drop(['OBJECTID','EASTING','NORTHING'], axis = 1, inplace=True) # don't need these (Easting is OS equivalent of x)
pdTFL_tubes.rename(columns = {'x':'latitude', 'y':'longitude'}, inplace = True)  # we do need these : y is latitude

pdTFL_tubes = pdTFL_tubes.dropna()  # drop the 200 or so observations that are not part of a London Underground line (i.e. Rail, Overground etc)
pdTFL_tubes[0:5]
```
